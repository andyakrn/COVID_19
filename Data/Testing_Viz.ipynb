```python
import pandas as pd
import plotly.express as px
```

```python
dff = pd.read_csv('State data/states.csv')
```

```python
dff.info()
```

```python
df = dff[['state', 'positive', 'negative', 'pending', 'hospitalized', 'death', 'total']].copy()
```

```python
df.fillna(0, inplace = True)
```

```python
df['quarantined'] = df['positive'] - (df['hospitalized'] + df['death'])
```

```python
df
```

```python
states = pd.read_excel('State data/nst-est2019-01.xlsx', header = 3)
```

```python
st_df = states[5:58].dropna()
```

```python
st_pop = st_df[['Unnamed: 0', 2019]].copy()
```

```python
st_pop.columns =  ['State', 'Population (est.)']
```

```python
st_pop['State'] = st_pop['State'].str.replace('.', '')
```

```python
st_pop
```

```python
st_abv = pd.read_csv('State data/us_states.csv', header = None)
```

```python
st_abv.columns = ['All_caps', 'Capitalized', 'Abbrev']
```

```python
st_pop = st_abv.merge(st_pop, left_on = 'Capitalized', right_on = 'State', how = 'right')
```

```python
st_pop = st_pop[['Abbrev', 'State', 'Population (est.)']].copy()
```

```python
st_pop
```

```python
st_pop['Abbrev'].loc[st_pop['State'] == 'District of Columbia'] = 'DC'
st_pop['Abbrev'].loc[st_pop['State'] == 'Puerto Rico'] = 'PR'
```

```python
states_df = df.merge(st_pop, left_on = 'state', right_on = 'Abbrev', how = 'left')
```

```python
states_df
```

```python
states_df['State'].loc[states_df['state'] == 'AS'] = 'American Samoa'
states_df['State'].loc[states_df['state'] == 'GU'] = 'Guam'
states_df['State'].loc[states_df['state'] == 'MP'] = 'Northern Marianas'
states_df['State'].loc[states_df['state'] == 'VI'] = 'Virgin Islands'
```

```python
states_df.drop(columns = ['state', 'Abbrev'], inplace = True)
```

```python
states_df
```

```python
states_df = states_df[['State', 'Population (est.)','positive', 'negative', 'pending', 'hospitalized', 'death', 'total',
                     'isolating']]
```

```python
states_df['Population (est.)'].loc[states_df['State'] == 'American Samoa'] = 55689.0
states_df['Population (est.)'].loc[states_df['State'] == 'Guam'] = 167772.0
states_df['Population (est.)'].loc[states_df['State'] == 'Northern Marianas'] = 53883.0
states_df['Population (est.)'].loc[states_df['State'] == 'Virgin Islands'] = 104578.0
```

```python
states_df
```

```python
states_df['untested'] = states_df['Population (est.)'] - states_df['total']
```

```python
states_df
```

```python
plt_df = states_df[['State', 'untested','pending', 'negative', 'hospitalized', 'death', 'quarantined']].copy()
```

```python
plt_df = plt_df.set_index('State').transpose().reset_index()
```

```python
plt_df
```

```python
plt_df['tested'] = ['untested', 'tested', 'tested', 'tested', 'tested', 'tested']
plt_df['results'] = [ None , 'pending', 'negative', 'positive', 'positive', 'positive']
plt_df['actions'] = [ None, None , None, 'hospitalized', 'death', 'quarantined']
```

```python
pct_df = states_df[['State','Population (est.)','untested','pending', 'negative', 'hospitalized', 'death', 'quarantined']].copy()
```

```python
pct_df['untested'] = round(((pct_df['untested'] / pct_df['Population (est.)']) * 100), 4)
```

```python
pct_df['pending'] = round(((pct_df['pending'] / pct_df['Population (est.)']) * 100), 4)
pct_df['negative'] = round(((pct_df['negative'] / pct_df['Population (est.)']) * 100), 4)
pct_df['hospitalized'] = round(((pct_df['hospitalized'] / pct_df['Population (est.)']) * 100), 4)
pct_df['death'] = round(((pct_df['death'] / pct_df['Population (est.)']) * 100), 4)
pct_df['quarantined'] = round(((pct_df['quarantined'] / pct_df['Population (est.)']) * 100), 4)
```

```python
pct_df
```

```python
plt_pct_df = pct_df.set_index('State').transpose().reset_index()
```

```python
plt_pct_df.drop(0, inplace = True)
```

```python
plt_pct_df['tested'] = ['untested', 'tested', 'tested', 'tested', 'tested', 'tested']
plt_pct_df['results'] = [ None , 'pending', 'negative', 'positive', 'positive', 'positive']
plt_pct_df['actions'] = [ None, None , None, 'hospitalized', 'death', 'quarantined']
```

```python
fig = px.sunburst(plt_pct_df, path = ['tested', 'results', 'actions'], values= 'Louisiana', color= 'index',
                  color_discrete_map= {'untested' :'lightgray'})
fig.show()
```

```python
fig = px.treemap(plt_df, path = ['tested', 'results', 'actions'], values= 'New York', color= 'index',
                 color_discrete_map= {'untested' :'lightgray'})
fig.show()
```
